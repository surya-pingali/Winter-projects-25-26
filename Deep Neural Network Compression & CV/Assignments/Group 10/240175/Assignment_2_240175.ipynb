````python
def gray_histogram(gray_img):
  #creating a zero array of size 255, where we will store the count
  hist = np.zeros(256, dtype = int)

  #dimension of the image
  rows, cols = gray_img.shape

  for i in range(rows):
    for j in range(cols):
      pixelValue = gray_img[i,j]
      hist[pixelValue] += 1

  return hist

histogram_data = gray_histogram(img_gray)
#plotting
plt.figure(figsize=(10, 5))
plt.title("Grayscale Histogram")
plt.xlabel("Pixel Intensity (0-255)")
plt.ylabel("Frequency (Count)")
plt.bar(range(256), histogram_data, width=1.0, color='blue', edgecolor='blue')
plt.xlim([-0.5, 255.5]) # Set limits to cover all bins continuously
plt.show()
````

````python
def rgb_histogram(img_rgb):
  rows, cols, _ = img_rgb.shape
  channels =[0,1,2]
  colors = ['r', 'g', 'b']
  hist = np.zeros((3,256))
  for c in channels:
    for i in range(rows):
      for j in range(cols):
        pixel_value = img_rgb[i,j,c]
        hist[c,pixel_value] += 1
  return hist

histogram_data = rgb_histogram(img_rgb)

# Plotting the RGB histogram
plt.figure(figsize=(12, 6))
colors = ['red', 'green', 'blue']
labels = ['Red Channel', 'Green Channel', 'Blue Channel']

for i in range(3):
    plt.plot(range(256), histogram_data[i], color=colors[i], label=labels[i])

plt.title("RGB Histogram")
plt.xlabel("Pixel Intensity (0-255)")
plt.ylabel("Frequency (Count)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim([-0.5, 255.5])
plt.show()
````

#### RGB to HSV ( only Numpy ofc )
Write a function <code>rgb_to_hsv</code> that takes a 0-255 ( not normalised ) RGB image and returns an HSV image ( also not normalised ).

**Note :** Research for the formulas, maybe ask some AI to explain you the algo and formulas ( NOT to give you the code, seriously youre not gonna learn anything if you do that )

The output should be in a format that is compatible with OpenCV cvtColor. Read the documentation to find out any Hue scaling etc.
Finally, use your function and test it by using cv2.cvtColor to convert your function output into RGB again and display it. Your image will appear the same after reconversion into RGB.

````python
def rgb_to_hsv(image_rgb):

    #Here we convert to float to avoid overflow or underflow during math calculations
    img_float = image_rgb.astype(np.float32)

    R = img_float[..., 0]
    G = img_float[..., 1]
    B = img_float[..., 2]

    # Calculation of V

    cmax = np.max(img_float, axis=2)
    cmin = np.min(img_float, axis=2)
    delta = cmax - cmin

    V = cmax

    # S caluclation
    S = np.zeros_like(cmax)

    non_zero_cmax = cmax != 0
    S[non_zero_cmax] = (delta[non_zero_cmax] / cmax[non_zero_cmax]) * 255

    # Hue(H) calculation

    H = np.zeros_like(cmax)
    delta_mask = delta != 0

    # Where Red is Max
    mask_r = (cmax == R) & delta_mask
    H[mask_r] = 30 * (((G[mask_r] - B[mask_r]) / delta[mask_r]) % 6)

    # Where Green is Max
    mask_g = (cmax == G) & delta_mask
    H[mask_g] = 30 * (2 + (B[mask_g] - R[mask_g]) / delta[mask_g])

    # Where Blue is Max
    mask_b = (cmax == B) & delta_mask
    H[mask_b] = 30 * (4 + (R[mask_b] - G[mask_b]) / delta[mask_b])

    H = np.round(H).astype(np.uint8)
    S = np.round(S).astype(np.uint8)
    V = np.round(V).astype(np.uint8)

    # Stacking channels back into an image
    return np.dstack((H, S, V))


hsv_func = rgb_to_hsv(img_rgb)

#converting to rgb from hsv using opencv
reconstructed_rgb = cv2.cvtColor(hsv_func, cv2.COLOR_HSV2RGB)

plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.title("Original RGB")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("HSV")
plt.imshow(hsv_func)
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Reconstructed RGB (using cv2)")
plt.imshow(reconstructed_rgb)
plt.axis('off')

plt.show()
````

````python

def billa_laser_eye(image):

    # I have Created a copy of image
    img = image.copy()

    # --- Modification 1: Saturation Boost ---

    hsv_img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
    h, s, v = cv2.split(hsv_img)

    s = np.clip(s.astype(np.float32) * 1.5, 0, 255).astype(np.uint8)


    hsv_new = cv2.merge([h, s, v])
    saturated_img = cv2.cvtColor(hsv_new, cv2.COLOR_HSV2RGB)


    final_img = cv2.convertScaleAbs(saturated_img, alpha=1.1, beta=10)

    return final_img

filtered_img_1 = billa_laser_eye(img_rgb)

plt.figure(figsize=(5, 5))
plt.imshow(filtered_img_1)
plt.title("Modification 1")
plt.axis('off')
plt.show()

#I chose these modifications because the cat's most striking feature is its yellow eyes. The saturation makes the eye of the cat to glow more and contrast adjustment defines the texture of grey fur.
````

````python
import cv2
import numpy as np

def garam_billa(image):

    img = image.copy()


    img_float = img.astype(np.float32)

    # Boost Red channel (Index 0 in RGB)
    img_float[:, :, 0] = img_float[:, :, 0] + 30
    # Boost Green channel slightly (Index 1)
    img_float[:, :, 1] = img_float[:, :, 1] + 15
    # Reduce Blue channel (Index 2)
    img_float[:, :, 2] = img_float[:, :, 2] - 10


    tinted_img = np.clip(img_float, 0, 255).astype(np.uint8)


    gamma = 0.8
    invGamma = 1.0 / gamma

    table = np.array([((i / 255.0) ** invGamma) * 255
                      for i in np.arange(0, 256)]).astype("uint8")


    final_img = cv2.LUT(tinted_img, table)

    return final_img

filtered_img_2 = garam_billa(img_rgb)

plt.figure(figsize=(5, 5))
plt.imshow(filtered_img_2)
plt.title("Modification 2")
plt.axis('off')
plt.show()

# This modifcation makes the image look more warmer.
````

Use Numpy only to keep track of the brightest pixel and then scale the entire image with the factor you find. Display the original and output image.

````python
def white_patch_balance(img_rgb):

    img_float = img_rgb.astype(float)


    max_vals = np.max(img_float, axis=(0, 1))


    scale_factors = np.where(max_vals == 0, 1, 255.0 / max_vals)


    balanced_img = img_float * scale_factors


    balanced_img = np.clip(balanced_img, 0, 255).astype(np.uint8)

    return balanced_img

wb_image = white_patch_balance(img_rgb)


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(wb_image)
plt.title("White Patch Balanced")
plt.axis('off')

plt.tight_layout()
plt.show()
````
