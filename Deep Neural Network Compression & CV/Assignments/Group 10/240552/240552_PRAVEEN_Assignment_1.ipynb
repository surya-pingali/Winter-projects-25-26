````python
# code for C1
row,col=gray2.shape
crow,ccol=row//2,col//2
LPF_mask_1=np.zeros((row,col),dtype=np.uint8)
LPF_mask_2=np.zeros((row,col),dtype=np.uint8)
LPF_mask_3=np.zeros((row,col),dtype=np.uint8)
r1=10
r2=50
r3=100# here i considered the radii 10,50,100 to show the clear difference in images due to change in radii 
for i in range(crow-r1,crow+r1):
    for j in range(ccol-r1,ccol+r1):
        if (i-crow)**2+(j-ccol)**2<=(r1)**2:
            LPF_mask_1[i,j]=1
for i in range(crow-r2,crow+r2):
    for j in range(ccol-r2,ccol+r2):
        if (i-crow)**2+(j-ccol)**2<=(r2)**2:
            LPF_mask_2[i,j]=1
for i in range(crow-r3,crow+r3):
    for j in range(ccol-r3,ccol+r3):
        if (i-crow)**2+(j-ccol)**2<=(r3)**2:
            LPF_mask_3[i,j]=1
# the mask of required radius r is created
f_LPF_1=fshift*LPF_mask_1
ishift_LPF_1=np.fft.ifftshift(f_LPF_1)
img_back_1=np.fft.ifft2(ishift_LPF_1)
img_back_LPF_1=np.abs(img_back_1)
f_LPF_2=fshift*LPF_mask_2
ishift_LPF_2=np.fft.ifftshift(f_LPF_2)
img_back_2=np.fft.ifft2(ishift_LPF_2)
img_back_LPF_2=np.abs(img_back_2)
f_LPF_3=fshift*LPF_mask_3
ishift_LPF_3=np.fft.ifftshift(f_LPF_3)
img_back_3=np.fft.ifft2(ishift_LPF_3)
img_back_LPF_3=np.abs(img_back_3)
plt.figure(figsize=(24,6))
plt.subplot(1,4,1)
plt.imshow(gray2,cmap='gray')
plt.title("grayscale image")
plt.axis("off")
plt.subplot(1,4,2)
plt.imshow(img_back_LPF_1,cmap='gray')
plt.title("grayscale image after mask r1")
plt.axis("off")
plt.subplot(1,4,3)
plt.imshow(img_back_LPF_2,cmap='gray')
plt.title("grayscale image after mask r2")
plt.axis("off")
plt.subplot(1,4,4)
plt.imshow(img_back_LPF_3,cmap='gray')
plt.title("grayscale image after mask r3")
plt.axis("off")
plt.tight_layout()
plt.show()
''' form here we can observe that images become blur when radii
decreases this is because when we apply the np.fft.fftshift then all the low frequency terms of FFT i.e the values of low K are transferred to the middle 
forcing the high frequecny terms to the corners when we apply the LPF mask which is bascially 1's in the middle and zeros in the entire remaning region
 when convoluted to the fshit array all the low k terms are highleted decreasing this high k terms this leads to the erasing the sharp edges finally 
 leading to blurred image. Now as the radius decreased the 1's region becomes smaller incerasing the dominance of low k terms leading to even blurred image[ here k is the term in DTF
 formulae,which is directly proporsitional to frequency]'''
pass
````
