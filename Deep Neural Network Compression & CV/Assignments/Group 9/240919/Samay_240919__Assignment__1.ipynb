````python
radii = [10,30,80]
plt.figure(figsize=(15, 5))

rows, cols = gray.shape
crow, ccol = rows//2, cols//2

for i, radius in enumerate(radii):
  # create the mask
  mask_lpf = np.zeros((rows, cols), np.uint8)
  cv2.circle(mask_lpf, (ccol, crow), radius, 1, -1)

  # apply to fshift
  f_lpf = fshift*mask_lpf

  # Reconstruct using ifftshift + ifft2
  ishift_lpf = np.fft.ifftshift(f_lpf)
  img_back_lpf = np.fft.ifft2(ishift_lpf)
  img_back_lpf = np.abs(img_back_lpf)

  plt.subplot(1, 3, i+1)
  plt.imshow(img_back_lpf, cmap='gray')
  plt.title(f'LPF Radius: {radius}')
  plt.axis('off')

plt.show()

````

````python
radii = [10,30,80]
plt.figure(figsize=(15, 5))

rows, cols = gray.shape
crow, ccol = rows//2, cols//2

for i, radius in enumerate(radii):
  # create the mask
  mask_lpf = np.zeros((rows, cols), np.uint8)
  cv2.circle(mask_lpf, (ccol, crow), radius, 1, -1)

  mask_hpf = 1- mask_lpf
  f_hpf = fshift*mask_hpf

  # inverse fft for high pass
  ishift_hpf = np.fft.ifftshift(f_hpf)
  img_back_hpf = np.fft.ifft2(ishift_hpf)
  img_back_hpf = np.abs(img_back_hpf)

  plt.subplot(1, 3, i+1)
  plt.imshow(img_back_hpf, cmap='gray')
  plt.title(f'hPF Radius: {radius}')
  plt.axis('off')

plt.show()

````

````python
dist =  25
top_left = (ccol - dist, crow - dist)
bottom_right = (ccol + dist, crow + dist)
# create the mask
mask_square_lpf = np.zeros((rows, cols), np.uint8)
cv2.rectangle(mask_square_lpf, top_left, bottom_right, 1, -1)

# apply to fshift
f_sq_lpf = fshift*mask_square_lpf

# Reconstruct using ifftshift + ifft2
ishift_sq_lpf = np.fft.ifftshift(f_sq_lpf)
img_back_sq_lpf = np.fft.ifft2(ishift_sq_lpf)
img_back_sq_lpf = np.abs(img_back_sq_lpf)

plt.imshow(img_back_sq_lpf, cmap='gray')
plt.title(f'LPF size: {dist}')
plt.axis('off')

plt.show()


````

````python
phase_only = fshift / np.abs(fshift)
ishift_phase_only = np.fft.ifftshift(phase_only)
img_back_phase_only = np.fft.ifft2(ishift_phase_only)
img_back_phase_only = np.abs(img_back_phase_only)

plt.imshow(img_back_phase_only, cmap='gray')
plt.axis('off')

plt.show()

#fshift / np.abs(fshift): Returns a Complex Number . This format is ready to be fed directly back into the Inverse FFT function.
#np.angle(fshift): Returns a Real Number.
#Crucial Difference: we cannot put just an angle (real number) into ifft2. The Inverse FFT expects complex numbers.
#If we used np.angle, we would first have to convert it back to complex form using Euler's formula: complex_data = 1 * np.exp(1j * angle_data).
````

````python
phase_only1 = np.angle(fshift)
ishift_phase_only1 = np.fft.ifftshift(phase_only1)
img_back_phase_only1 = np.fft.ifft2(ishift_phase_only1)
img_back_phase_only1 = np.abs(img_back_phase_only1)

plt.imshow(img_back_phase_only1, cmap='gray')
plt.axis('off')

plt.show()
````

````python
mag_only = np.abs(fshift)
ishift_mag_only = np.fft.ifftshift(mag_only)
img_back_mag_only = np.fft.ifft2(ishift_mag_only)
img_back_mag_only = np.abs(img_back_mag_only)

plt.imshow(img_back_mag_only, cmap='gray')
plt.axis('off')

plt.show()
# Without Log: If you display it directly (linear scale), the huge center value drowns out everything else. You would just see a black square with a white dot.

# With Log: It compresses the huge values and boosts the small values so our eyes can see the "starburst" pattern and the details of the frequencies.
````
