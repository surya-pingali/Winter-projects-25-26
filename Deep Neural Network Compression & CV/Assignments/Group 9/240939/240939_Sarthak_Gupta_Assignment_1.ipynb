````python
#Creating circular masks using cv2 instead of manually making using Numpy
rows,cols=gray_img.shape

mask_10=np.zeros((rows,cols),np.uint8)
cv2.circle(mask_10,(cols//2,rows//2),10,1,-1)

mask_30=np.zeros((rows,cols),np.uint8)
cv2.circle(mask_30,(cols//2,rows//2),30,1,-1)

mask_80=np.zeros((rows,cols),np.uint8)
cv2.circle(mask_80,(cols//2,rows//2),80,1,-1)

#Making function to apply fshift and then reconstruct, since we have to do this multiple times
def apply_filter(mask):
    fshift_filtered=fshift*mask
    f_ishift=np.fft.ifftshift(fshift_filtered)
    img_new=np.fft.ifft2(f_ishift)
    return np.abs(img_new)

# Reconstructed images for each radius
img_10 = apply_filter(mask_10)
img_30 = apply_filter(mask_30)
img_80 = apply_filter(mask_80)
````

````python
#Filtered images plotted using subplot. We can see that on decreasing radius of low pass filter, blur level increases.
#This is because sharp edges are stored in the higher frequency componenets. When we lower radius, we exclude even more higher frequency components, thus image loses sharp edge content and becomes blurred
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(img_10, cmap='gray')
plt.title("Radius 10")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(img_30, cmap='gray')
plt.title("Radius 30")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(img_80, cmap='gray')
plt.title("Radius 80")
plt.axis("off")

plt.show()
````

````python
#Part C2
#Creating HPF. I will use radius as 80
#Since only higher frequency components are present, we can clearly see the edges in the image, and blurred portions are not present (blacked out)
mask_hpf_80 = np.ones((rows, cols), np.uint8)
cv2.circle(mask_hpf_80, (cols//2, rows//2), 80, 0, -1)

img_hpf_80=apply_filter(mask_hpf_80)
plt.imshow(img_hpf_80,cmap='gray')
plt.axis("off")
````
