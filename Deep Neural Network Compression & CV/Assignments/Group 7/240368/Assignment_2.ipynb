````python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

filename = 'rcrc.jpg'
img_bgr = cv.imread(filename, cv.IMREAD_COLOR)

img_rgb = cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)

def rgb_to_hsv(img_rgb):
    img_float = img_rgb.astype(np.float32) / 255.0

    R = img_float[:, :, 0]
    G = img_float[:, :, 1]
    B = img_float[:, :, 2]

    Cmax = np.max(img_float, axis=-1)
    Cmin = np.min(img_float, axis=-1)
    Delta = Cmax - Cmin

    H = np.zeros_like(Cmax)

    mask_r = (Cmax == R) & (Delta > 0)
    mask_g = (Cmax == G) & (Delta > 0)
    mask_b = (Cmax == B) & (Delta > 0)

    H[mask_r] = 60 * (((G[mask_r] - B[mask_r]) / Delta[mask_r]) % 6)
    H[mask_g] = 60 * (((B[mask_g] - R[mask_g]) / Delta[mask_g]) + 2)
    H[mask_b] = 60 * (((R[mask_b] - G[mask_b]) / Delta[mask_b]) + 4)

    S = np.zeros_like(Cmax)
    non_zero_cmax = Cmax > 0
    S[non_zero_cmax] = Delta[non_zero_cmax] / Cmax[non_zero_cmax]

    V = Cmax

    H_cv = (H / 2).astype(np.uint8)
    S_cv = (S * 255).astype(np.uint8)
    V_cv = (V * 255).astype(np.uint8)

    return np.dstack((H_cv, S_cv, V_cv))

hsv_img_manual = rgb_to_hsv(img_rgb)

img_recovered = cv.cvtColor(hsv_img_manual, cv.COLOR_HSV2RGB)

plt.figure(figsize=(16, 10))

plt.subplot(2, 2, 1)
plt.imshow(hsv_img_manual[:, :, 0], cmap='hsv')
plt.title("Hue Channel (Colors Mapped)")
plt.colorbar(label='Hue Value (0-179)')
plt.axis("off")

plt.subplot(2, 2, 2)
plt.imshow(hsv_img_manual[:, :, 1], cmap='gray')
plt.title("Saturation Channel\n(White = Vivid, Black = Faded)")
plt.axis("off")

plt.subplot(2, 2, 3)
plt.imshow(hsv_img_manual[:, :, 2], cmap='gray')
plt.title("Value Channel\n(Brightness)")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.imshow(img_recovered)
plt.title("Recovered RGB Image")
plt.axis("off")

plt.tight_layout()
plt.show()
````

Write a function <code>white_patch_balance(img_rgb)</code> that takes a 0-255 RGB image and then returns a White Patch Method Balanced 0-255 RBG image.
````python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

filename = 'underwater.jpg'

img_bgr = cv.imread(filename, cv.IMREAD_COLOR)


img_rgb = cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)

def white_patch_balance(img_rgb):
    img_float = img_rgb.astype(np.float32)

    max_r = np.max(img_float[:, :, 0])
    max_g = np.max(img_float[:, :, 1])
    max_b = np.max(img_float[:, :, 2])

    scale_r = 255.0 / (max_r + 1e-6)
    scale_g = 255.0 / (max_g + 1e-6)
    scale_b = 255.0 / (max_b + 1e-6)

    img_float[:, :, 0] *= scale_r
    img_float[:, :, 1] *= scale_g
    img_float[:, :, 2] *= scale_b

    img_balanced = np.clip(img_float, 0, 255)

    return img_balanced.astype(np.uint8)


balanced_img = white_patch_balance(img_rgb)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(balanced_img)
plt.title("White Patch Balanced")
plt.axis("off")

plt.show()
````
