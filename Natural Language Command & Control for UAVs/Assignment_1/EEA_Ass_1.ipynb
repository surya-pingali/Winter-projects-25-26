```python
def state_of_users(server_string):

  state_user = {}
  user_list = server_string.split("; ")

  for user_action_pair in user_list:
    if not user_action_pair: # Handle potential empty strings from split
      continue
    user_name, action = user_action_pair.split(": ")

    if action == "Login":
      state_user[user_name] = "Online"
    elif action == "Logout":
      state_user[user_name] = "Offline"
    # If other actions are possible, handle them here

  return state_user # Return the dictionary after processing all users

# These lines should be outside the function definition
server_string = "User1: Login; User2: Login; User1: Logout; User3: Login; User2: Logout"

final_state = state_of_users(server_string)

print(f"Server Log Input: {server_string}")
print(f"Final User States: {final_state}")
```

```python
## WE will create a Safe Calulator
def safe_calculator():
  print("Enter q to quit")
  while True:
    try:
         num1_str= input("Enter the first number: ")
         if num1_str.lower() == 'q':
                print("Exiting calculator.")
                break
         operator = input("Enter the operator: ")
         num2_str = input("Enter the second number: ")

         Num1 = float(num1_str)
         Num2 = float(num2_str)

         result = None

         if operator == "+":
           result = Num1 + Num2
         elif operator == "-":
           result = Num1 - Num2
         elif operator == "*":
           result = Num1 * Num2
         elif operator == "/":
             result = Num1 / Num2
         else:
          print("Invalid operator")
          continue

    except ZeroDivisionError:
         print("Error: Division by zero is not allowed.")
    except ValueError:
           print("Invalid input. Please enter a number.")
    else:
      print(f"{result}")
    finally:
      print("Execution attempt complete")


safe_calculator()
```

```python
class Book:

  def __init__(self, title, author):
    self.title = title
    self.author = author
    self.is_checked_out= False

  def __str__(self):
    status = "Checked Out" if self.is_checked_out else "Available"
    return f'"{self.title}" by {self.author} ({status})'

class Library:

  def __init__(self):
    self.books = []

  def add_book(self, book):
    self.books.append(book)

  def check_out_book(self, title):
    for book in self.books:
      if book.title == title :
        if book.is_checked_out:
          print(f'"{title}" is already checked out.')
          return # Book found, already checked out, exit.
        else:
          book.is_checked_out = True
          print(f'"{title}" has been checked out.')
          return # Book found, checked out, exit.
    print(f'"{title}" not found in the library.') # Book not found after loop, print message.

  def return_book(self, title):
    for book in self.books:
      if book.title== title:
        if book.is_checked_out:
          book.is_checked_out = False
          print(f'Returned : "{title}"')
          return # Book found and returned, exit.
        else:
          print(f'"{title}" is not currently checked out.')
          return # Book found but not checked out, exit.
    print(f'"{title}" not found in the library.') # Book not found after loop, print message.

library = Library()
book1 = Book("The Great Gatsby", "F. Scott Fitzgerald")
book2 = Book("To Kill a Mockingbird", "Harper Lee")
book3 = Book("1984", "George Orwell")

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

library.check_out_book("The Great Gatsby")
library.check_out_book("To Kill a Mockingbird")
library.check_out_book("1984")
library.check_out_book("The Catcher in the Rye")

library.return_book("The Great Gatsby")
library.return_book("To Kill a Mockingbird")

print("\nLibrary Books:")
for book in library.books:
  print(book)
```

```python
class TimeDuration:

    def __init__(self, hours=0, minutes=0):

        total_minutes = hours * 60 + minutes
        self.hours = total_minutes // 60
        self.minutes = total_minutes % 60


    def __add__(self, other):
        total_mins_self = self.hours * 60 + self.minutes
        total_mins_other = other.hours * 60 + other.minutes

        total_combined_minutes = total_mins_self + total_mins_other
        new_hours = total_combined_minutes // 60
        new_minutes = total_combined_minutes % 60


        return TimeDuration(hours=new_hours, minutes=new_minutes)


    def __str__(self):
        return f"{self.hours}H:{self.minutes}M"

t_norm = TimeDuration(hours=2, minutes=70)
print(f"Normalization Test (2h 70m): {t_norm}")


t1 = TimeDuration(hours=2, minutes=45)
t2 = TimeDuration(hours=1, minutes=30)

t_sum = t1 + t2
print(f"Addition Test (t1 + t2): {t_sum}")
```
