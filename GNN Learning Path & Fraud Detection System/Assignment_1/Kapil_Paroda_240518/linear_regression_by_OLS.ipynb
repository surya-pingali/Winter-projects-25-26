```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
square_footage = np.array([
    1100, 1400, 1425, 1550, 1600, 1700, 1750, 1800, 1875, 2000,
    2100, 2250, 2300, 2400, 2450, 2600, 2800, 2900, 3000, 3150
])
```

```python
price = np.array([
    199000, 245000, 230000, 215000, 280000, 295000, 345000, 315000, 325000, 360000,
    350000, 385000, 390000, 425000, 415000, 455000, 465000, 495000, 510000, 545000
])
```

```python
X = square_footage
```

```python
Y = price
```

```python
n = len(X)
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)
```

```python
numerator = np.sum((X -mean_x)*(Y- mean_y))
```

```python
denominator = np.sum((X - mean_x)**2)
```

```python
m = numerator / denominator
```

```python
b = mean_y - (m * mean_x)
```

```python
print("Calculated Parameters :")
print(f"m = {m}")
print(f"b = {b}")
```

```python
def predict_price(sqft, m, b):
    return m * sqft + b
```

```python
sqft_to_predict = 2500
```

```python
predicted_price = predict_price(sqft_to_predict,m,b)
```

```python
print(f"The predicted price for a house with {sqft_to_predict} square feet is: ${predicted_price:.2f}")
```

```python
Y_pred = predict_price(X,m,b)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Actual house prices')
plt.plot(X, Y_pred, color='red', label=f'Best fit line: y = {m:.2f}x + {b:.2f}')
plt.title('House price prediction using OLS linear regression')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
