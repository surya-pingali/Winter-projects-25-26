```python
with open("housing_prices.csv", "r") as f:
  lines=f.readlines()
```

```python
X=[]
Y=[]
```

```python
for line in lines[1:]: #header skipped
  clean_line=line.strip()
  area, price = clean_line.split(",")
  X.append(float(area))
  Y.append(float(price))
```

```python
print(X)
print(Y)
```

```python
N = len(X)

# initialise parameters
m = 0
b = 0

alpha = 1e-7
epochs = 50000
```

```python
for epoch in range(epochs):
  y_pred = [m * x + b for x in X]
  dm = (2.0 / N) * sum((y_pred[i] - Y[i]) * X[i] for i in range(N))
  db = (2.0 / N) * sum((y_pred[i] - Y[i]) for i in range(N))

  #updating
  m -= alpha * dm
  b -= alpha * db
```

```python
print("Final parameters from Gradient Descent:")
print("m =", m)
print("b =", b)
```

```python
price_2500 = m * 2500 + b
print(f"Price for 2500 sq feet= {price_2500:.2f}")
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, Y, label="Datapoints")

best_fity = [m * x + b for x in X]
plt.plot(X, best_fity, color='red', label="gradient descent best fit line")

plt.xlabel("Area in sq ft")
plt.ylabel("Price")
plt.title("Linear regression")
plt.show()
```
