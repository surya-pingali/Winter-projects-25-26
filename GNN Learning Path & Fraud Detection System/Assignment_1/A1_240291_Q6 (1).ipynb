```python
import numpy as np
import matplotlib.pyplot as plt

square_feet = [1100, 1400, 1425, 1550, 1600, 1700, 1750, 1800, 1875,
               2000, 2100, 2250, 2300, 2400, 2450, 2600, 2800, 2900, 3000, 3150, 3300 ]


prices = [199000, 245000, 230000, 215000, 280000, 295000, 345000,315000,
          325000, 360000, 350000, 385000, 390000, 425000, 415000,455000,
          465000, 495000, 510000, 545000, 570000]

x = np.array(square_feet, dtype=float)
y = np.array(prices, dtype=float)

x_mean = np.mean(x)
y_mean = np.mean(y)

m = np.random.rand()
b = np.random.rand()

alpha = 1e-8

eps = 1e-5

m_update = 0
b_update = 0
M = len(y)
i=0
while ((abs(m_update-m) > eps) or (abs(b_update-b) >eps)) and i<100000 :
    if i!=0:
        m = m_update
        b = b_update
    i+=1
    y_pred = m*x + b
    dm = 1/M*np.sum((y_pred-y)*x)
    db = 1/M*np.sum(y_pred-y)
    m_update = m - alpha*dm
    b_update = b - alpha*db

sqft_to_predict = 2500
predicted_price = m * sqft_to_predict + b

print(f"Slope (m): {m:.2f}")
print(f"Intercept (b): {b:.2f}")
print(f"Predicted price of a 2500 sq ft house: Rs.{predicted_price:.2f}")

plt.scatter(x, y, label='Data Points')
x_line = np.linspace(min(x), max(x))
y_line = m * x_line + b
plt.plot(x_line, y_line, label='Best Fit Line', linewidth=2)

plt.xlabel("Square Footage")
plt.ylabel("Price (Rs.)")
plt.title("Housing Price Prediction")
plt.grid(True)
plt.legend()
plt.show()
```
