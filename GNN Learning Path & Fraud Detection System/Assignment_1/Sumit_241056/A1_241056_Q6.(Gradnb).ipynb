```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/housing_prices - housing_prices.csv')
x_data = df.iloc[:, 0].values
y_data = df.iloc[:, 1].values
```

```python
def gradient_descent(x, y, learning_rate, iterations):
    m = 0.0
    b = 0.0
    n = len(y)

    for _ in range(iterations):
        y_pred = m * x + b
        dm = (-2/n) * np.sum(x * (y - y_pred))
        db = (-2/n) * np.sum(y - y_pred)
        m = m - learning_rate * dm
        b = b - learning_rate * db

    return b, m
```

```python
learning_rate = 1e-7
iterations = 100000
b, m = gradient_descent(x_data, y_data, learning_rate, iterations)
```

```python
prediction_sqft = 2500
predicted_price = m * prediction_sqft + b
print(f"Gradient Descent Predicted price for {prediction_sqft} sq ft: ${predicted_price:.2f}")
plt.figure(figsize=(10, 6))
plt.scatter(x_data, y_data, color='green', label='Actual Data')
x_line = np.linspace(x_data.min(), x_data.max(), 100)
y_line = m * x_line + b

plt.plot(x_line, y_line, color='orange', label=f'GD Line: y={m:.2f}x + {b:.2f}')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.title('Housing Prices: Gradient Descent Method ')
plt.legend()
plt.grid(True)
plt.show()
```
