```python


import urllib.request
import matplotlib.pyplot as plt


sheet_url = "https://docs.google.com/spreadsheets/d/1au3HI9TCGs8dy_YIpGVPJizmI_-eeyXB0jdVcGCEXss/export?format=csv"

response = urllib.request.urlopen(sheet_url)
csv_data = response.read().decode('utf-8')

lines = csv_data.split('\n')
square_footage = []
price = []

for i in range(1, len(lines)):
    if lines[i].strip():  # Skip empty lines
        parts = lines[i].split(',')
        square_footage.append(float(parts[0]))
        price.append(float(parts[1]))

print("Data parsed successfully!")
print("Number of data points:", len(square_footage))
print("Square Footage (first 5):", square_footage[:5])
print("Price (first 5):", price[:5])

sum_x = 0
for x in square_footage:
    sum_x = sum_x + x
mean_x = sum_x / len(square_footage)

sum_y = 0
for y in price:
    sum_y = sum_y + y
mean_y = sum_y / len(price)

print("Mean Square Footage:", mean_x)
print("Mean Price:", mean_y)

numerator = 0
for i in range(len(square_footage)):
    numerator = numerator + (square_footage[i] - mean_x) * (price[i] - mean_y)

denominator = 0
for i in range(len(square_footage)):
    denominator = denominator + (square_footage[i] - mean_x) ** 2

m = numerator / denominator

print("Slope (m):", m)

b = mean_y - m * mean_x

print("Y-Intercept (b):", b)
print("\nLine of Best Fit Equation:")
print("Price = {:.2f} * SquareFootage + {:.2f}".format(m, b))

predictions = []
for x in square_footage:
    predictions.append(m * x + b)

tss = 0
for y in price:
    tss = tss + (y - mean_y) ** 2

rss = 0
for i in range(len(price)):
    rss = rss + (price[i] - predictions[i]) ** 2

r_squared = 1 - (rss / tss)

print("\nModel Performance:")
print("R-squared:", r_squared)
print("This means the model explains {:.1f}% of the variance in house prices".format(r_squared * 100))

house_size = 2500

predicted_price = m * house_size + b

print("\n" + "=" * 50)
print("PREDICTION RESULT")
print("=" * 50)
print("Predicted price for a {} sq ft house: ${:.2f}".format(house_size, predicted_price))
print("=" * 50)



plt.figure(figsize=(12, 6))

plt.scatter(square_footage, price, color='blue', label='Actual Data', s=50, alpha=0.6)

# Generate points for the best fit line
x_line = [min(square_footage), max(square_footage)]
y_line = [m * x + b for x in x_line]

# Plot the best fit line
plt.plot(x_line, y_line, color='red', label='Best Fit Line', linewidth=2)

# Plot the prediction point
plt.scatter([house_size], [predicted_price], color='green', s=150,
            label='Prediction (2500 sq ft)', marker='*', zorder=5, edgecolor='black')

# Add labels and title
plt.xlabel('Square Footage', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.title('House Price Prediction using OLS Method\nRÂ² = {:.4f}'.format(r_squared),
          fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

# Display the plot
plt.tight_layout()
plt.show()

print("\nEquation: Price = {:.2f} * SquareFootage + {:.2f}".format(m, b))
```

```python
import urllib.request
import matplotlib.pyplot as plt

sheet_url = "https://docs.google.com/spreadsheets/d/1au3HI9TCGs8dy_YIpGVPJizmI_-eeyXB0jdVcGCEXss/export?format=csv"

response = urllib.request.urlopen(sheet_url)
csv_data = response.read().decode('utf-8')

lines = csv_data.split('\n')

square_footage = []
price = []

for i in range(1, len(lines)):
    if lines[i].strip():  # Skip empty lines
        parts = lines[i].split(',')
        square_footage.append(float(parts[0]))
        price.append(float(parts[1]))

mean_x = sum(square_footage) / len(square_footage)
mean_y = sum(price) / len(price)

std_x = (sum((x - mean_x)**2 for x in square_footage) / len(square_footage))**0.5
std_y = (sum((y - mean_y)**2 for y in price) / len(price))**0.5

X_standardized = [(x - mean_x) / std_x for x in square_footage]
y_standardized = [(y - mean_y) / std_y for y in price]


m = 0.0
b = 0.0

learning_rate = 0.001
iterations = 10000


n = len(X_standardized)

print("Initial parameters:")
print("m (slope) =", m)
print("b (intercept) =", b)
print("Learning rate =", learning_rate)
print("Iterations =", iterations)


cost_history = []
m = 0
b = 0
n = len(X_standardized)

for iteration in range(iterations):

    # compute y_pred AFTER the latest m,b
    y_pred = [m * X_standardized[i] + b for i in range(n)]

    # gradients (using clean sums instead of noisy loop accumulation)
    dm = -(2/n) * sum((y_standardized[i] - y_pred[i]) * X_standardized[i] for i in range(n))
    db = -(2/n) * sum((y_standardized[i] - y_pred[i]) for i in range(n))

    # update
    m = m - learning_rate * dm
    b = b - learning_rate * db

    # compute cost
    cost = sum((y_pred[i] - y_standardized[i])**2 for i in range(n)) / (2*n)
    cost_history.append(cost)

    if (iteration + 1) % 500 == 0:
        print(f"Iteration {iteration+1}: Cost={cost:.6f}, m={m:.4f}, b={b:.4f}")

print("\nGradient Descent Completed!")
print(f"Final standardized m = {m}, b = {b}")


m_original = m * (std_y / std_x)
b_original = mean_y - m_original * mean_x

print("\nOriginal scale parameters:")
print(f"Slope (m) = {m_original:.4f}")
print(f"Intercept (b) = {b_original:.4f}")



# Calculate predictions for all data points
predictions = []
for x in square_footage:
    predictions.append(m_original * x + b_original)

# Calculate Total Sum of Squares (TSS)
tss = 0
for y in price:
    tss = tss + (y - mean_y) ** 2

# Calculate Residual Sum of Squares (RSS)
rss = 0
for i in range(len(price)):
    rss = rss + (price[i] - predictions[i]) ** 2

# Calculate R-squared
r_squared = 1 - (rss / tss)

print("\nModel Performance:")
print("R-squared:", r_squared)


# House size to predict
house_size = 2500

# Prediction formula: y = mx + b
predicted_price = m_original * house_size + b_original

print("\n" + "=" * 50)
print("PREDICTION RESULT")
print("=" * 50)
print("Predicted price for a {} sq ft house: ${:.2f}".format(house_size, predicted_price))
print("=" * 50)

plt.figure(figsize=(12, 5))
plt.plot(cost_history, color='blue', linewidth=2)
plt.xlabel('Iteration', fontsize=12)
plt.ylabel('Cost (MSE)', fontsize=12)
plt.title('Cost Function Convergence', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 6))

plt.scatter(square_footage, price, color='blue', label='Actual Data', s=50, alpha=0.6)

x_line = [min(square_footage), max(square_footage)]
y_line = [m_original * x + b_original for x in x_line]

plt.plot(x_line, y_line, color='red', label='Best Fit Line', linewidth=2)
plt.scatter([house_size], [predicted_price], color='green', s=150,
            label='Prediction (2500 sq ft)', marker='*', zorder=5, edgecolor='black')

plt.xlabel('Square Footage', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.title('House Price Prediction using Gradient Descent\nRÂ² = {:.4f}'.format(r_squared),
          fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nFinal Equation: Price = {:.2f} * SquareFootage + {:.2f}".format(m_original, b_original))
```
