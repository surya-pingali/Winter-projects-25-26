```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
square_feet=[]
prices=[]
with open("housing_prices - housing_prices.csv","r") as f:
    reader=csv.reader(f)
    next(reader)
    for row in reader:
      sq=float(row[0])
      price=float(row[1])
      square_feet.append(sq)
      prices.append(price)
```

```python
x=np.array(square_feet)
y=np.array(prices)
```

```python
x_mean=np.mean(x)
y_mean=np.mean(y)
m=np.sum((x-x_mean)*(y-y_mean))
n=np.sum((x-x_mean)**2)
slope=m/n
b=y_mean-slope*x_mean
```

```python
def predict_price(sqft):
  return slope*sqft+b
```

```python
pred_2500=predict_price(2500)
print(f"\nPredicted price for 2500 sq_ft = ${pred_2500:.2f}")
```

```python
plt.scatter(x,y,label="Data Points")
plt.plot(x,m*x+b, label="Best-Fit-Line")
plt.xlabel("Square Feet")
plt.ylabel("Price")
plt.title("Housing Prices- Best Fit Line")
plt.legend
plt.show()
```
