```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = []
y= []
```

```python
import urllib.request

url = "https://docs.google.com/spreadsheets/d/1au3HI9TCGs8dy_YIpGVPJizmI_-eeyXB0jdVcGCEXss/export?format=csv"

response = urllib.request.urlopen(url)
data = response.read().decode('utf-8')
lines = data.split("\n")

x = []
y = []
for line in lines[1:]:
    line = line.strip()
    features = line.split(",")
    sq = features[0]
    price = features[1]

    x.append(float(sq))
    y.append(float(price))
```

```python
x = np.array(x)
y = np.array(y)

n= len(x)
m = 0
b = 0
L = 0.00000005
iterations = 1000000
n = len(x)

for i in range(iterations):
    y_pred = m*x + b

    dm = (1/n) * np.sum((y_pred - y) * x)
    db = (1/n) * np.sum(y_pred - y)

    m = m - L * dm
    b = b - L * db

predicted_price = m*2500+b
print(f"The predicted price of 2500 sqft is : {predicted_price:.2f}")
```

```python
plt.scatter(x, y, color="blue", label="Actual Data")
plt.plot(x, m*x + b, color="red", label="Best Fit Line")
plt.xlabel("Sq. feets")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```
