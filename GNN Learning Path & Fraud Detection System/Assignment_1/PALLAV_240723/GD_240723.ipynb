```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1100, 1400, 1425, 1550, 1600, 1700, 1750, 1800, 1875, 2000, 2100, 2250, 2300, 2400, 2450, 2600, 2800, 2900, 3000, 3150, 3300])
y = np.array([199000, 245000, 230000, 215000, 280000, 295000, 345000, 315000, 325000, 360000, 350000, 385000, 390000, 425000, 415000, 455000, 465000, 495000, 510000, 545000, 570000])
m_samples = len(x)
# i am also doing ML from coursera Andrew Ng , i have used this template as it was taught in that course and changed somewhere as required
```

```python
def compute_gradient(x, y, w, b):
    m = len(x)
    dj_dw = 0.0
    dj_db = 0.0

    for i in range(m):
        f_wb = w * x[i] + b
        error = f_wb - y[i]

        dj_dw += error * x[i]
        dj_db += error

    dj_dw /= m
    dj_db /= m

    return dj_dw, dj_db
```

```python
def gradient_descent(x, y, w_init, b_init, alpha, epochs):
    w = w_init
    b = b_init


    for epoch in range(epochs):
        dj_dw, dj_db = compute_gradient(x, y, w, b)

        w = w - alpha * dj_dw
        b = b - alpha * dj_db



    return w, b
```

```python

# parameters for GD
w_init = 0.0   # slope
b_init = 0.0   # intercept
alpha = 1e-7   # learning rate
epochs = 10000 # no of iterations

w, b = gradient_descent(x, y, w_init, b_init, alpha, epochs)

print("Final slope w:", w)
print("Final intercept b:", b)
```

```python
value = w * 2500 + b
print(f"Predicted price for 2500 sq ft: {value:.2f}")
```

```python
plt.scatter(x, y, label="Data")
plt.plot(x, w * x + b, color="orange", label="GD Best Fit")
plt.xlabel("Square Footage")
plt.ylabel("Price")
plt.title("Gradient Descent")
plt.legend()
plt.show()
```
