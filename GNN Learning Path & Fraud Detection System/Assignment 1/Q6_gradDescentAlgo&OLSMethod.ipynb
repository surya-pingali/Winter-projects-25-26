```python
import numpy as np
import matplotlib.pyplot as plt

# Data
raw_data = """
1100,199000
1400,245000
1425,230000
1550,215000
1600,280000
1700,295000
1750,345000
1800,315000
1875,325000
2000,360000
2100,350000
2250,385000
2300,390000
2400,425000
2450,415000
2600,455000
2800,465000
2900,495000
3000,510000
3150,545000
3300,570000
"""

X = []
Y = []

for line in raw_data.strip().split("\n"):
    x, y = line.split(",")
    X.append(float(x))
    Y.append(float(y))

X = np.array(X)
Y = np.array(Y)

# Gradient Descent Implementation

m = 0.0
b = 0.0
learning_rate = 1e-8
itr = 500000
n = len(X)

for _ in range(itr):
    y_pred = m * X + b
    dm = (-2/n) * np.sum(X * (Y - y_pred))
    db = (-2/n) * np.sum(Y - y_pred)

    m -= learning_rate * dm
    b -= learning_rate * db

# Prediction for 2500 sqft

x_test = 2500
y_test = m * x_test + b

print("Predicted Price for 2500 sqft using Gradient Descent: ₹{:.2f}".format(y_test))

# Plot Best Fit Line

x_line = np.linspace(min(X), max(X), 100)
y_line = m * x_line + b

plt.scatter(X, Y)
plt.plot(x_line, y_line)
plt.xlabel("Square Footage")
plt.ylabel("Price")
plt.title("House Price Prediction using Gradient Descent")
plt.grid(True)
plt.show()
```

```python
#OLS Method
import numpy as np
import matplotlib.pyplot as plt

# Sample data
raw_data = """
1100,199000
1400,245000
1425,230000
1550,215000
1600,280000
1700,295000
1750,345000
1800,315000
1875,325000
2000,360000
2100,350000
2250,385000
2300,390000
2400,425000
2450,415000
2600,455000
2800,465000
2900,495000
3000,510000
3150,545000
3300,570000
"""

X = []
Y = []

for line in raw_data.strip().split("\n"):
    x, y = line.split(",")
    X.append(float(x))
    Y.append(float(y))

X = np.array(X)
Y = np.array(Y)

# 1) ORDINARY LEAST SQUARES (MANUAL FORMULA)

x_mean = np.mean(X)
y_mean = np.mean(Y)

numerator = np.sum((X - x_mean) * (Y - y_mean))
denominator = np.sum((X - x_mean) ** 2)

m_ols = numerator / denominator
b_ols = y_mean - m_ols * x_mean

# Prediction using OLS
x_test = 2500
y_test_ols = m_ols * x_test + b_ols

print("OLS Prediction for 2500 sqft: ₹{:.2f}".format(y_test_ols))

# PLOTTING BEST FIT LINE

x_line = np.linspace(min(X), max(X), 100)

y_line_ols = m_ols * x_line + b_ols

plt.scatter(X, Y)
plt.plot(x_line, y_line_ols, label="OLS Best Fit")
plt.xlabel("Square Footage")
plt.ylabel("Price")
plt.title("House Price Prediction using OLS")
plt.legend()
plt.grid(True)
plt.show()
```
