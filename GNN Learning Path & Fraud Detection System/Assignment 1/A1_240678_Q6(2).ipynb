```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sqf=[1100, 1400, 1425, 1550, 1600, 1700, 1750, 1800, 1875, 2000, 2100, 2250, 2300, 2400, 2450, 2600, 2800, 2900, 3000, 3150, 3300]
price=[199000, 245000, 230000, 215000, 280000, 295000, 345000, 315000, 325000, 360000, 350000, 385000, 390000, 425000, 415000, 455000, 465000, 495000, 510000, 545000, 570000]

arr1=np.array(sqf)
arr2=np.array(price)
n=len(arr1)

m=0.0
b=0.0
a= 0.0000001
nooftimes=1000000

for i in range(nooftimes):
    pred=m*arr1+b
    error=arr2-pred
    dwrtm=(-2/n)*np.sum(arr1*error)
    dwrtb=(-2/n)*np.sum(error)
    m=m-a*dwrtm
    b=b-a*dwrtb

print(f"Slope:{m:.2f}")
print(f"Intercept:{b:.2f}")

def predict_price(sq_feet):
    return m*sq_feet+b

predicted_price = predict_price(2500)
print(f"Predicted price:{predicted_price:.2f}")

plt.scatter(arr1, arr2, color='blue', label='Actual data')

x_line = np.linspace(min(arr1)-100, max(arr1)+100, 500)
y_line = m * x_line + b
plt.plot(x_line, y_line, color='red', label='Best fit line')

plt.xlabel("Square Feet")
plt.ylabel("Price ($)")
plt.title("House Price vs Square Feet")
plt.legend()
plt.grid(True)
plt.show()
```
